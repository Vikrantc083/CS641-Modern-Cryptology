```python
# Permute function to rearrange the bits 
def permutation_box_des(k, arr, n):
    permutation = ""
    i = 0
    while i < n:
        permutation = permutation + k[arr[i] - 1]
        i += 1
    return permutation
# shifting the bits towards left by nth shifts 
def shift(k, n):
    shifted = k[n:] + k[0:n]
    return shifted
```

```python
# Function to read ciphertexts
def read_ciphertext(file_name):
    ciphertext_des1 = open(file_name).read().split("\n")
    hexcipher_des1 = []
    for c in ciphertext_des1:
        h = []
        if len(c) == 16:
            for i in range(16):
                h.extend(mapping[c[i]])
            hexcipher_des1.extend([h])
    return hexcipher_des1

# Function to perform inverse final permutation
def inverse_final_permutation(hexcipher_des1):
    invres_des1 = []
    for c in hexcipher_des1:
        temp = []
        for i in range(64):
            temp.extend([c[Reverse_Final_Permutation[i] - 1]])
        invres_des1.extend([temp])
    return invres_des1

# Function to compute XOR of ciphertext pairs
def compute_xor(invres_des1):
    resxor_des1 = []
    i = 0
    while i < len(invres_des1) // 2:
        resxor_des1.extend([list(np.bitwise_xor(invres_des1[2 * i], invres_des1[2 * i + 1]))])
        i += 1
    return resxor_des1

# Function to perform expansion of right side block of Round 5
def expand_right_block(invres_des1):
    expanded_des1 = []
    for i in range(len(invres_des1)):
        temp = []
        for j in range(48):
            temp.extend([invres_des1[i][expansion_box[j] - 1]])
        expanded_des1.extend([temp])
    return expanded_des1

# Function to compute S-box input
def compute_sbox_input(expanded_des1):
    sin_des1 = []
    i = 0
    while i < len(expanded_des1) // 2:
        sin_des1.extend([list(np.bitwise_xor(expanded_des1[2 * i], expanded_des1[2 * i + 1]))])
        i += 1
    return sin_des1

# Function to perform XOR of L5 and R6
def xor_L5_R6(resxor_des1):
    L5 = [0, 0, 0, 0, 0, 1] + [0] * 26
    inxor = []
    for i in resxor_des1:
        inxor.extend([list(np.bitwise_xor(i[32:64], L5))])
    return inxor

# Function to compute XOR of S-box outputs
def compute_sbox_output(inxor):
    sxor_des1 = []
    for i in inxor:
        temp = []
        for j in range(32):
            temp.extend([i[Inverse_Permutation_BOX[j] - 1]])
        sxor_des1.extend([temp])
    return sxor_des1

# Function to find keys corresponding to Round 6 for the given characteristic
def find_keys(sin_des1, sxor_des1, expanded_des1):
    keys = np.zeros((8, 64))
    i = 0
    while i < len(sin_des1):
        if sin_des1[i] == "":
            i += 1
            continue
        j = 0
        while j < 8:
            inx = int(''.join([str(s) for s in sin_des1[i][j * 6:j * 6 + 6]]), 2)
            outx = int(''.join([str(s) for s in sxor_des1[i][j * 4:j * 4 + 4]]), 2)
            inp = int(''.join([str(s) for s in expanded_des1[2 * i + 1][j * 6:j * 6 + 6]]), 2)
            k = 0
            while k < 64:
                a = bin(k)[2:].zfill(6)
                b = bin(k ^ inx)[2:].zfill(6)
                if outx == sbox[j][int(a[0]) * 2 + int(a[5])][int(a[4]) + 2 * int(a[3]) + int(a[2]) * 4 + int(a[1]) * 8] ^ sbox[j][int(b[0]) * 2 + int(b[5])][int(b[4]) + 2 * int(b[3]) + int(b[2]) * 4 + int(b[1]) * 8]:
                    keys[j][k ^ inp] += 1
                k += 1
            j += 1
        i += 1
    return keys


hexcipher_des1 = read_ciphertext('ciphertext1.txt')
invres_des1 = inverse_final_permutation(hexcipher_des1)
resxor_des1 = compute_xor(invres_des1)
expanded_des1 = expand_right_block(invres_des1)
sin_des1 = compute_sbox_input(expanded_des1)
inxor = xor_L5_R6(resxor_des1)
sxor_des1 = compute_sbox_output(inxor)
keys = find_keys(sin_des1, sxor_des1, expanded_des1)
print(keys)


```

```python
# Define the S-box key
sb_key = "101101111011XXXXXX000111100101110010001110111101"
#Putting Sbox 3 as XXXXXX as it is never zero
# Initialize the key with 'X's
key = ['X']*56

# Update the key based on the S-box key
i = 0
while i < 48:
    key[Permuted_Choice_2[i]-1] = sb_key[i]
    i += 1

# Perform key schedule generation
i = 0
while i < 6:
    j = 0
    while j < shift_table[i]:
        x = key[27]
        y = key[55]
        k = 27
        while k > 0:
            key[k] = key[k-1]
            key[k+28] = key[k+27]
            k -= 1
        key[0] = x
        key[28] = y
        j += 1
    i += 1

# Print the generated key
print(''.join(key))
miskey = ''.join(key)
```

```python
poskey_des = []
binlist_des = []
i = 0
while i < 2**14:
    x = str(bin(i)[2:])
    binlist_des.append(('0')*(14-len(x)) + x)
    i += 1

for i in binlist_des:
    j = 0
    tempkey = list(miskey)
    k = 0
    while k < len(tempkey):
        if tempkey[k] == 'X':
            tempkey[k] = i[j]
            j += 1
        k += 1
    poskey_des.append(''.join(tempkey))

def roundkey_des(k, rno):
    left = k[0:28]
    right = k[28:56]

    binkey = []

    i = 0
    while i < rno:
        left = shift(left, shift_table[i])
        right = shift(right, shift_table[i])

        key = left + right

        binkey.extend([permutation_box_des(key, key_comp, 48)])
        i += 1

    return binkey

def encryption_des(mess, key, rno):
    mess = permutation_box_des(mess, initial_permutaion_table, 64)

    left = mess[:32]
    right = mess[32:]

    i = 0
    while i < rno:
        expmess = permutation_box_des(right, expansion_box, 48)
        inxor = str(bin(np.bitwise_xor(int(expmess, 2), int(key[i], 2)))[2:])
        if len(inxor) != 48:
            inxor = ('0'*(48-len(inxor))) + inxor
        sout = ''
        j = 0
        while j < 8:
            temp = (bin(sbox[j][int(inxor[j*6] + inxor[j*6+5], 2)][int(inxor[j*6+1:j*6+5], 2)])[2:])
            sout += ('0'*(4-len(temp)) + temp)
            j += 1
        sout = permutation_box_des(sout, sboxper, 32)

        roundxor = str(bin(np.bitwise_xor(int(left, 2), int(sout, 2)))[2:])
        if len(roundxor) != 32:
            roundxor = ('0'*(32-len(roundxor))) + roundxor
        left = roundxor

        if i != 5:
            t = left
            left = right
            right = t

        i += 1

    outmess = left + right
    cipher = permutation_box_des(outmess, final_permutation_box, 64)

    return cipher
```
